# Introduction

Le jeu de données étudié (SIGN Language mnist : https://www.kaggle.com/datamunge/sign-language-mnist) est composé d'images étiquetées représentant des lettres dans la langue des signes. 

L'objectif de ce projet est de faire de la classification d'images.

Les images sont étiquetées dans des entiers de 0 à 25, représentant les lettres dans l'ordre alphabétique. Il n'y a aucun cas pour J = 9 ainsi que pour Z = 25, car ces deux lettres sont représentées par des mouvements dans la langue des signes.
On dispose d'un dataset d'entrainement (27000 images) et d'un dataset de test (7000 images).
Les images sont au format standard de MNIST, à savoir de taille 28*28.

J'ai réalisé ce projet avec google colab.



# Importation et lecture des fichiers

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from sklearn.preprocessing import StandardScaler
```

```python
#Lecture des datasets d'entrainement et de test

train = pd.read_csv('Data/sign_mnist_test.csv')
test = pd.read_csv('Data/sign_mnist_test.csv')
```

```python
#Affichage du début du dataset d'entrainement pour voir la forme

train.head()
```

Remarque : Chaque image étiquetée est ici représentée par un vecteur de taille 785. 
Le premier entier représente la classe correspondante (1=A, 2=B, etc).
Les 784 suivants sont les pixels de l'image, de gauche à droite puis de haut en bas.

```python
lettres = ['A','B','C','D','E','F','G','H','I','K','L','M','N','O','P','Q','R','S','T','U','V','W','X','Y'] #Sans le J et le Z
```

Ci-dessus on retrouve les différentes lettres présentes dans le dataset, sachant que J et Z n'y sont pas car nécéssitant un mouvement

```python
X_train = train.drop("label", axis=1)
Y_train = train["label"]

Xt = np.array(X_train)

Yt = np.array(Y_train)
Xt[0].shape
```

```python
#Affichage des premières images
plt.figure()
plt.imshow(Xt[0].reshape(28,28),'gray')
plt.title(lettres[int(Yt[0])])
plt.figure()
plt.imshow(Xt[1].reshape(28,28),'gray')
plt.title(lettres[int(Yt[1])])
plt.figure()
plt.imshow(Xt[2].reshape(28,28),'gray')
plt.title(lettres[int(Yt[2])])
```

# Création du modèle

```python
Xt = Xt/255
Xt = Xt.reshape((-1,28,28)) #Ici, j'ai converti les images stockées dans des vecteurs de dimension (784,1) en (1,28,28)
```

```python
model = tf.keras.models.Sequential() #On fait un modèle séquentiel
```

```python
model.add(tf.keras.layers.Flatten(input_shape=[28,28]))
```

Remarque : Le fait de convertir les images en (1,28,28) puis d'effectuer un flatten n'est par inutile comme on pourrait le penser à première vue  : cela m'a permis d'éviter des erreurs de dimensions.

Je normalise ensuite les données

```python
model.add(tf.keras.layers.LayerNormalization())
```

Je crée ensuite un premier Layer de 256 neurones, puis un second de 128 neurones. Cette configuration se prête généralement bien à la classification d'images.

```python
model.add(tf.keras.layers.Dense(256, activation="relu"))
```

```python
model.add(tf.keras.layers.Dense(128,activation="relu"))
```

J'ai choisi de mettre 26 sorties possibles, correspondant aux 26 lettres de l'alphabet (avec Y=9 et Z=25 qui seront nuls).

```python
model.add(tf.keras.layers.Dense(26,activation="softmax"))
```

L'activation "softmax" fait que la somme de la valeur de tous les neurones vaut 1. C'est alors équivalent à une distribution de probabilités.

```python
first_output = model.predict(Xt[0:1]) #Prédiction fausse car le modèle n'a pas encore été entraîné.
```

```python
model.summary()
```

```python
model.compile(
    loss='sparse_categorical_crossentropy', 
    optimizer="sgd", #Descente de gradient stocastique
    metrics=["accuracy"]
)
```

# Entrainement du modèle

J'ai utilisé 70% du jeu de données pour l'entrainement et 3 pourcents pour la validation.
Le paramètre batch_size est par défaut à 32.

```python
history = model.fit(Xt,Yt,epochs=10,validation_split=0.3)
```

On voit que la précision converge très vite : de 54% à la fin du premier cycle, on est presque à 100% après l'époque 5.

Remarques : La normalisation des données a eu un rôle important : en effet, la précision plafonnait à 82% sans normaliser les données.

# Prédiction et mesure de la fiabilité sur le jeu de données de test

```python
X_test = test.drop("label", axis=1)
Y_test = test["label"]

Xtest = np.array(X_test)

Ytest = np.array(Y_test)
```

```python
output_tests = model.predict(Xtest)
```

```python
output_tests.shape
```

```python
nb_classification_justes = 0
nb_classification_fausses = 0
i = 0

for e in output_tests:
  vmax =e.max()
  prediction = int(np.where(e == vmax)[0][0])
  if (prediction == Ytest[i]):
    nb_classification_justes+=1
  else:
    nb_classification_fausses+=1
  i+=1

print("Nombre de classifications justes : ",nb_classification_justes, "\n")
print("Nombre de classifications fausses : ", nb_classification_fausses, "\n")
print("Pourcentage de classifications justes :",100*nb_classification_justes/len(output_tests),"%")
```

Les essais sur le jeu de données de test nous donnent une fiabilité du modèle de 100%. Cependant, il faut prendre du recul avec ce score, car les images du dataset sont toutes prises dans un contexte similaire.
